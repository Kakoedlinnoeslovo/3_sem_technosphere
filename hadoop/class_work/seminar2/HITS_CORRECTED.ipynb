```python
with open(fname) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.split() for x in content] 
```

```python
Hes = set([x[0] for x in content])
Aes = set([x[1] for x in content])
```

```python
from collections import Counter
```

```python
Hes_set = Counter()
Aes_set = Counter()
```

```python
for line in content:
    Aes_set[line[1]] = 1
    Hes_set[line[0]] = 1
```

```python
for i in range(10):
    for line in content:
        Aes_set[line[1]] += Hes_set[line[0]]
    for line in content:
        Hes_set[line[0]] += Aes_set[line[1]]
```
